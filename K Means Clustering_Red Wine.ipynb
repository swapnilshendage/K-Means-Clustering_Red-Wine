```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
wine_df = pd.read_csv('winequality-red.csv',sep=';')
wine_df.head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(wine_df.corr(),annot=True)
```

```python
wine_df = wine_df.drop('quality',axis=1)
wine_df.columns
```

```python
wss=[]
for i in range(1,10):
    kmeans = KMeans(n_clusters=i,init='k-means++', random_state=42)
    kmeans.fit(wine_df)
    wss.append(kmeans.inertia_)
    
plt.plot(range(1,10),wss)
plt.title('The Elblow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distance')
plt.show()
```

```python
for i in range(2,10):
    kmeans = KMeans(n_clusters=i,max_iter=100)
    kmeans.fit(wine_df)
    score = silhouette_score(wine_df,kmeans.labels_)
    print('For clusters {}, the silhouette score is {}'.format(i,score))
```

```python
silhouette_coefficients = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i,max_iter=100)
    kmeans.fit(wine_df)
    score = silhouette_score(wine_df,kmeans.labels_)
    silhouette_coefficients.append(score)
    
plt.plot(range(2,10),silhouette_coefficients)
plt.xticks(range(2,10))
plt.xlabel('Number of clusters')
plt.ylabel('silhouette coefficients')
plt.show()
```

```python
pca = PCA()
X = pca.fit_transform(wine_df)
```

```python
kmeans = KMeans(n_clusters=3)
label = kmeans.fit_predict(X)
unique_labels = np.unique(label)
```

```python
for i in unique_labels:
    plt.scatter(X[label==i,0],X[label==i,1],label=i,s=40)

plt.legend()
plt.title('Wine Groups')
plt.show()
```
